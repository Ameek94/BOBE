```python
# create external likelihood

ndim = 6
# example multivariate gaussian in 6 dimensions 
def logp(x, mean, cov = np.eye(ndim)):
    diff = np.array(x-mean).T
    res = np.linalg.solve(cov,diff)
    res = -0.5 * np.dot(diff.T,res)
    res-= 0.5*ndim*(np.log(2*np.pi) + np.log(np.linalg.det(cov)))
    return res

mean = np.zeros(ndim)
x = np.random.uniform(size=(6))
logp(x,mean)
param_bounds= np.array(ndim*[[-5,5]]).T
gaussian = external_loglike(logposterior=logp,ndim=ndim, param_bounds=param_bounds)

```
